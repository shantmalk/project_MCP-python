# Import dependencies

```python
import pyodbc
import pandas as pd
```

## Connect to database
* Make sure to download the correct ODBC driver for access 64 bit: https://www.microsoft.com/en-us/download/details.aspx?id=54920
* Make sure that you are connected to uci-vpn otherwise polaris will not be available

```python
db_type = 'Microsoft Access Driver (*.mdb, *.accdb)'
db_path = r"C:\Users\Admin\Desktop\db_CONFIRM.accdb"
pyodbc_str = (
    r'DRIVER={{{DB_TYPE}}};'
    r'DBQ={DB_PATH};'
    ).format(DB_TYPE=db_type, DB_PATH=db_path)
```

```python
conn = pyodbc.connect(pyodbc_str)
```

## Create our first SQL query

```python
sql_query = '''
SELECT *
FROM tblconfirmCONFIRM
'''
```

```python
pd.read_sql_query(sql_query, conn)
```
